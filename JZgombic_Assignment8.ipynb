The first step in our program is to import all the required tools that we will need including **pandas**, **numpy**, **matplotlib** and **seaborn**.
    
We also need to include the line *%matplotlib inline*

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

In this block of code, we are just setting the default size of tables that are returned. In this case, I chose 10 columns wide by 40 rows long.

```python
pd.set_option('display.max_columns', 10)
pd.set_option('display.max_rows', 40)
```

In this block of code, we are readings the data into pythion as well as naming the individual columns using **names = []** and removing excessive whitespace using **sep = "\s+"**.

```python
mpg = pd.read_csv('auto-mpg.data', 
                  sep = "\s+", 
                  names =['MPG', 
                          'Cylinders', 
                          'Displacement', 
                          'Horsepower', 
                          'Weight', 
                          'Acceleration', 
                          'Model Year', 
                          'Origin',
                          'Car Name'], 
                  header= None)
```

In this block of code, we are using **.replace()** to change all the ? to NaN, all the 1s to USA, all the 2s to Asia and all the 3s to Europe.

```python
mpg.replace({'?':np.nan, 1:'USA', 2:'Asia', 3:'Europe'})
```

We are now checking the data types of all the columns using **.dtypes** and we see that **Horsepower**, which we may use for calculations is an object so we will convert that to a number in the next step.

```python
mpg.dtypes
```

Using **.to_isnumeric**, we will convert **Horsepower** to **float64** and using **.dtypes** once again, we see that we are successful.

```python
mpg['Horsepower'] = pd.to_numeric(mpg['Horsepower'], errors ='coerce')
mpg.dtypes
```

First, we are looking in all the columns for just **Cylinders** or **Car Name**. We will then use **.groupby()** to group the data by **Cylinders** and using **count()**, we will get the number of vehicles under each **Cylinder** category. We will then go ahead and use **.rename()** to rename the **Car Name** column to **No. of Vehicles** for a correct representation. Finally, we will sort the values by the highest number of **Cylinders** using **.sort_values()**.

```python
cylinders = mpg[[col for col in mpg.columns if 'Cylinders' in col or 'Car Name' in col]]
grouped = cylinders.groupby('Cylinders').count()
grouped.rename(columns={'Car Name': 'No. of Vehicles'}, inplace=True)
grouped.sort_values(by=['No. of Vehicles'], ascending = False)
```

We are now setting the default size of any graphs we use. I have chosen 15 wide by 10 long fo a larger graph.

```python
plt.rcParams["figure.figsize"] = (15,10)
```

Using **sns.setstyle**, we will set the grid to be white and using **.plot()**, we will plot a bar graph using the grouped **Cylinder data** from the previous step.

```python
sns.set_style('whitegrid')
plotCylinder = grouped.plot(kind='bar', color='red')
```

Using **plt.scatter()**, we create a scatter type graph showing the relationship between the **Horsepower** of a vehicle and its **Weight**. As you can clearly see, the heavier the vehicle, the more **Horsepower** we get. There are some exceptions, however, the overall statement is true.

```python
plt.scatter(mpg.Weight, mpg.Horsepower)
```

Using **seaborn**, we are plotting a scatter graph using the **MPG** data and **Horsepower** of vehicles. As we saw in the previous example when **Horsepower** was related to **Weight**, we are seeing almost the same distribution here when it comes to these data points. The more **Horsepower** we get, the lower the **MPG**. It may not be true for all data but the overall distribution shows us that it is the norm.

```python
sns.set(color_codes=True)
sns.regplot(x=mpg.Horsepower, y=mpg.MPG, data=mpg, color='g')
```
